create a dashoard to analyze the unemployement rate of each indian state by region #use sunburst plot

```python
import pandas as pd
import plotly.express as px  # Import Plotly Express

# Assuming you already have your 'data' DataFrame

Unemployment = data[["State", "Region", "Estimated Unmeployement Rate"]]

figure = px.sunburst(
    Unemployment,
    path=["Region", "State"],  # Corrected column name "State" instead of "States"
    values="Estimated Unmeployement Rate",
    width=700,
    height=700,
    color_continuous_scale="RdYlGn",  # Adjust the color scale as needed
    title="Unemployment Rate in India"
)

figure.show()
```
